$\textbf{1.- ¿Cuál es el vector de estado de un sistema compuesto por 3 qubits , estando los qubits individuales(e independientes) en los estados  $|+\rangle $ , $|+\rangle $,$| -\rangle $,? Luego hacer las mediciones parciales para el qubit central, repetir lo mismo con qiskit.}$

Tenemos $$|\psi\rangle = |+\rangle \otimes |+\rangle \otimes |-\rangle = |+\rangle|+\rangle|-\rangle$$
Escribiendolo en la base computacional tenemos $$= \frac{1}{2^{3/2}}(|0\rangle + |1\rangle)(|0\rangle + |1\rangle)(|0\rangle - |1\rangle)$$ $$=\frac{1}{2^{2/3}}(|00\rangle + |01\rangle + |10\rangle + |11\rangle)(|0\rangle - |1\rangle)$$ $$|\psi\rangle=\frac{1}{2^{3/2}}(|000\rangle - |001\rangle + |010\rangle - |011\rangle + |100\rangle - |101\rangle + |110\rangle - |111\rangle)$$
Como nos interesa en obtener la amplitud de probabilidad de medir el qubit de en medio lo aislamos como sigue: $$|\psi\rangle=\frac{1}{2^{3/2}}(|0\rangle + |1\rangle)\otimes |0\rangle \otimes (|0\rangle - |1\rangle) + \frac{1}{2^{3/2}}(|0\rangle + |1\rangle)\otimes |1\rangle \otimes (|0\rangle - |1\rangle)$$

Entonces la probabilidad de que el qubit de en medio este en el estado $|0\rangle$ al medirlo es $$p_0 = |\frac{1}{2^{3/2}}(|0\rangle + |1\rangle)|^2||0\rangle|^2||0\rangle - |1\rangle|^2 = \frac{1}{2^3}(\langle 0| + \langle 1|)(|0\rangle + |1\rangle)\langle 0| 0\rangle(\langle 0| - \langle 1|)(|0\rangle - |1\rangle)$$ $$=\frac{1}{2^3}(\langle 0| 0\rangle + \langle 1| 1\rangle)(\langle 0| 0\rangle + \langle 1| 1\rangle) = \frac{1}{2^3} /times 2 \times 2$$ Por lo que $$p_0 = \frac{1}{2}$$
Para la probabilidad de encontrar el qubit de en medio en el estado $|1\rangle$ tenemos $$p_1 = |\frac{1}{2^{3/2}}(|0\rangle + |1\rangle)|^2||1\rangle|^2||0\rangle - |1\rangle|^2 = p_0 = \frac{1}{2}$$

$\textbf{3.- ¿Cómo se escribiría en notación de Dirac una compuerta CU3 desarrollalo e implementalo en qiskit para $\theta = \pi$, $\phi = \frac{\pi}{2}$ y $\lambda = \frac{7 \pi}{2}$}$

Como queremos que cuando el qubit $q_0$ se encuentre en el estado $|0\rangle$ la compuerta no actue sobre $q_1$ entonces tenemos el primer termino como el producto tensorial $I\otimes|0\rangle\langle0|$. Como queremos que la compuerta U3 actue sobre $q_1$ cuando $q_0$ se encuentre en $|1\rangle$ entonces tenemos el termino $U3(\theta,\phi,\lambda)\otimes|1\rangle\langle1|$ , por lo que $$CU3(\theta,\phi,\lambda) = I\otimes|0\rangle\langle0| + U3(\theta,\phi,\lambda)\otimes|1\rangle\langle1|$$

Ya que $$U3 = \begin{bmatrix} \cos{\frac{\theta}{2}} & -e^{i\lambda}\sin{\frac{\theta}{2}} \\ e^{i\phi}\sin{\frac{\theta}{2}} & e^{i(\phi+\lambda)}\cos{\frac{\theta}{2}}
\end{bmatrix}$$

Podemos escribirla en las bases $|0\rangle$ y $|1\rangle$: $$U3 = |0\rangle \langle 0| \cos{\frac{\theta}{2}} - |0\rangle \langle 1| e^{i\lambda}\sin{\frac{\theta}{2}} + |1\rangle \langle 0| e^{i\phi}\sin{\frac{\theta}{2}} + |1\rangle \langle 1| e^{i(\phi + \lambda)}\cos{\frac{\theta}{2}}$$ Por lo que $$CU3 = I\otimes|0\rangle\langle0| + (|0\rangle \langle 0| \cos{\frac{\theta}{2}} - |0\rangle \langle 1| e^{i\lambda}\sin{\frac{\theta}{2}} + |1\rangle \langle 0| e^{i\phi}\sin{\frac{\theta}{2}} + |1\rangle \langle 1| e^{i(\phi + \lambda)}\cos{\frac{\theta}{2}})\otimes|1\rangle\langle1|$$
En este caso el operador de la izquierda en el producto tensorial actua sobre el qubit $q_1$ y el de la derecha sobre $q_0$ de acuerdo con la convencion que usa Qiskit.

Apliquemos esta compuerta para el caso en donde inicialmente tenemos $|q_0\rangle = |1\rangle$ y $|q_1\rangle = |0\rangle$ de tal manera que la compuerta U3 se vea implementada:

$$CU3|01\rangle = CU3 |0\rangle \otimes |1\rangle = I |0\rangle \otimes |0\rangle\langle0|1\rangle + (|0\rangle \langle 0|0\rangle \cos{\frac{\theta}{2}} - |0\rangle \langle 1|0\rangle e^{i\lambda}\sin{\frac{\theta}{2}} + |1\rangle \langle 0|0\rangle e^{i\phi}\sin{\frac{\theta}{2}} + |1\rangle \langle 1|0\rangle e^{i(\phi + \lambda)}\cos{\frac{\theta}{2}})\otimes|1\rangle\langle1|1\rangle$$

Usando la ortonormalidad de los estados: $$CU3|01\rangle = (\cos{\frac{\theta}{2}}|0\rangle + e^{i\phi}\sin{\frac{\theta}{2}}|1\rangle)|1\rangle$$
Para los valores $\theta = \pi$ y $\phi = \frac{\pi}{2}$: $$CU3|01\rangle =  e^{i\frac{\pi}{2}}\sin{\frac{\pi}{2}}|1\rangle|1\rangle = i|11\rangle$$ obtenemos el estado $|11\rangle$ desfasado.

Ahora lo implementamos en qiskit:

```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.circuit.library.standard_gates import U3Gate
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere
import numpy as np
```

```python
qr = QuantumRegister(2)
qc = QuantumCircuit(qr)
qc.initialize(Statevector.from_label("1"),0)
cu3_gate = U3Gate(np.pi,np.pi/2,7*np.pi/2).control(1)
qc.append(cu3_gate,qr)
qc.draw('mpl')
```

```python
plot_bloch_multivector(qc)
```

El circuito para el oraculo cuya funcion cumple con la tabla indicada arriba tiene la forma que se muestra en la siguiente imagen:



Verificacion:



La siguiente funcion implementa dicho oraculo:

```python
def f_balanceada(n: int):
    f_b = QuantumCircuit(n)

    f_b.cx(range(n-1),n-1)

    return f_b
```

```python
fb = f_balanceada(3)
fb.draw('mpl')
```

Implementamos el resto del circuito para verificar que esta funcion es balanceada:

```python
def q_c(function: QuantumCircuit):
    n = function.num_qubits
    qc1 = QuantumCircuit(n,n-1)
    
    qc1.x(n-1)
    qc1.barrier()

    for i in range(n):
        qc1.h(i)
    qc1.barrier()

    qc1.compose(function, inplace=True)
    qc1.barrier()

    for i in range(n-1):
        qc1.h(i)
    qc1.barrier()

    qc1.measure(range(n-1),range(n-1))

    return qc1
```

```python
from qiskit_aer import AerSimulator
from qiskit import transpile
```

```python
def algoritmo(fn: QuantumCircuit):
    backend = AerSimulator()
    qc_trans = transpile(fn,backend)
    result = backend.run(qc_trans,shots=1,memory=True).result()
    measurement = result.get_memory()

    if '1' in measurement[0]:
      print("la funcion es balanceada")
    else:
        print("la funcion es constante")
```

```python
qcf = q_c(fb)
qcf.draw('mpl')
```

```python
algoritmo(qcf)
```

$\textbf{7.- Implemente el algoritmo de Grover para determinar la máxima cantidad de aristas en un grafo de 5 nodos y 6 aristas:}$

\begin{align}
(0,3)\\

Los conjuntos que se deben elegir para obtener el mayor número de aristas correctas son {0,1,2} y {3,4}, en donde el mayor numero de aristas correctas es 6 como se muestra a continuacion:



Implementamos el algoritmo de acuerdo a esta configuracion:

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.circuit.library import XGate,ZGate
from qiskit.visualization import plot_histogram
from qiskit import execute, Aer
```

```python
def oracle_part(qc,qr):
    
    mcx1 = XGate().control(3) #para operar con compuertas controladas por 3 qubits control y 1 target.
    mcx2 = XGate().control(4) #para operar con compuertas controladas por 4 qubits control y 1 target.
    
    # como son seis aristas ponemos 6 qubits auxiliares para detectar las aristas correctas

    # ocuparemos 4 ultimos qubits auxiliares que nos digan la suma de aristas correctas

    # ocuparemos un ultimo qubit auxiliar que nos indicara si tenemos el maximo numero de aristas correctas
    
    #Los qubits del 0 al 4 son los nodos. Cada nodo es un qubit (n=5 en este caso).
    
    #Los qubits del 5 al 10 nos permiten reconocer si la arista es correcta (otros 6 qubits para verificar)
    #(i.e. si se da entre nodos de distinto color).
    
    #cada barrier es una arista definidas todas como arriba (0,3),(0,4),(1,3),(1,4),(2,3),(2,4)

    
    qc.cx(qr[0], qr[5])
    qc.cx(qr[3], qr[5])
    qc.barrier()
    qc.cx(qr[0], qr[6])
    qc.cx(qr[4], qr[6])
    qc.barrier()
    qc.cx(qr[1], qr[7])
    qc.cx(qr[3], qr[7])
    qc.barrier()
    qc.cx(qr[1], qr[8])
    qc.cx(qr[4], qr[8])
    qc.barrier()
    qc.cx(qr[2], qr[9])
    qc.cx(qr[3], qr[9])
    qc.barrier()
    qc.cx(qr[2], qr[10])
    qc.cx(qr[4], qr[10])
    qc.barrier()
    
    #Los qubits del 11 al 14 nos van a permitir contar la cantidad de aristas correctas
    #de tal forma que la suma será expresada en binario con cifras dadas por q_14, q_13 , q_12 y q_11.
    
    qc.barrier()    
    qc.cx(qr[5],qr[11])
    qc.barrier()
    
    qc.ccx(qr[6], qr[11], qr[12])
    qc.cx(qr[6],qr[11])
    qc.barrier()
    
    qc.ccx(qr[7], qr[11], qr[12])
    qc.cx(qr[7],qr[11])
    qc.barrier()
    
    qc.append(mcx1,[8,11,12,13])
    qc.ccx(qr[8], qr[11], qr[12])
    qc.cx(qr[8],qr[11])
    qc.barrier()
    
    qc.append(mcx1,[9,11,12,13])
    qc.ccx(qr[9], qr[11], qr[12])
    qc.cx(qr[9],qr[11])
    qc.barrier()

    qc.append(mcx2,[10,11,12,13,14])
    qc.append(mcx1,[10,11,12,13])
    qc.ccx(qr[10], qr[11], qr[12])
    qc.cx(qr[10],qr[11])
    qc.barrier()

    #como el 6 en binario es 0110 aplicamos una compuerta x en el qubit $q_11$ y otra en $q_14$
    qc.x(qr[11])
    qc.x(qr[14])
    qc.barrier()
    
    #El qubit 15 indicará cuando se haya encontrado un estado deseado
    qc.append(mcx2,[11,12,13,14,15])
    
    return qc
```

```python
qr=QuantumRegister(16)
qc=QuantumCircuit(qr)
oracle_part(qc,qr).draw()
```

De acuerdo a los conjuntos seleccionados, los estados deseados deben ser $|11000\rangle$ y $|00111\rangle$. Verificamos esto:

```python
# aplicamos compuerta z para transformar qubit q_15 a -1 e implementamos el proceso inverso con las compuertas

def oracle(qc,qr):
    
    qc=oracle_part(qc,qr)
    qc1=qc.inverse()
    qc.z(15)
    qc.compose(qc1,inplace=True)
    
    return qc
```

```python
# aplicamos el difusor

def grover_diffusion(qc,qr):
    mcz = ZGate().control(4)
    
    qc.barrier()
    qc.h(range(5))
    qc.x(range(5))
    qc.append(mcz,list(range(0,5))) 
    qc.x(range(5))
    qc.h(range(5))
    qc.barrier()
    return qc
```

```python
def grover(qc,qr):
    return grover_diffusion(oracle(qc,qr),qr) #aplicamos el oraculo y luego el difusor
```

```python
# implementamos un circuito cuantico como tal
qr=QuantumRegister(16) #registro cuantico
qc=QuantumCircuit(qr) # circuito cuantico
# creamos un registro total, usamos qra para tener otro registro y evitar error
qra=QuantumRegister(16)
cra=ClassicalRegister(5) # mediremos el valor de los cuatro nodos solamente (classical_n=4)
qa=QuantumCircuit(qra,cra)

qa.h(range(5)) # inicialmente se le debe aplicar un hadamard a los nodos para tener todos los estados posibles
qb=grover(qc,qr)
qa.compose(qb.compose(qb.compose(qb)),inplace=True) #aplicamos qb tres veces porque aqui k=3, i.e. aplicamos tres veces U_w U_f

qa.measure(range(5),cra)

qa.draw()
```

Simplificando $$|\psi_3\rangle = [-2(\frac{a}{2^n})^{\frac{3}{2}}) + 2 \frac{b}{2^n}\sqrt{\frac{a}{2^n}} + \sqrt{\frac{a}{2^n}}]|A\rangle + [2(\frac{b}{2^n})^{\frac{3}{2}}) - 2 \frac{a}{2^n}\sqrt{\frac{b}{2^n}} - \sqrt{\frac{b}{2^n}} ]|B\rangle$$
Simplificando mas $$|\psi_3\rangle = 2(\frac{a}{2^n})^{\frac{1}{2}}[-\frac{a}{2^n} + \frac{b}{2^n} + \frac{1}{2}]|A\rangle +  2(\frac{b}{2^n})^{\frac{1}{2}}[-\frac{a}{2^n} + \frac{b}{2^n} - \frac{1}{2}]|B\rangle$$

Sustituimos $\sqrt{\frac{a}{2^n}} = \sin{\theta}$ y $\sqrt{\frac{b}{2^n}} = \cos{\theta}$: $$|\psi_3\rangle = 2 \sin{\theta}[\cos^2{\theta} - \sin^2{\theta} + \frac{1}{2}]|A\rangle + 2 \cos{\theta}[\cos^2{\theta} - \sin^2{\theta} - \frac{1}{2}]|A\rangle$$ Usando $\cos^2{\theta} + \sin^2{\theta} = 1 $: $$|\psi_3\rangle = 2 \sin{\theta}[(1 - \sin^2{\theta}) - \sin^2{\theta} + \frac{1}{2}]|A\rangle + 2 \cos{\theta}[\cos^2{\theta} - (1 - \cos^2{\theta}) - \frac{1}{2}]|A\rangle$$ $$= 2 \sin{\theta}(\frac{3}{2} - 2 \sin^2{\theta})|A\rangle + 2 \cos{\theta}(2 \cos^2{\theta} - \frac{3}{2})|B\rangle$$ $$=(3 \sin{\theta} - 4\sin^3{\theta})|A\rangle + (4\cos^3{\theta} - 3\cos{\theta})|B\rangle$$
Usando las identidades $\sin{3 \theta} = 3 \sin{\theta} - 4\sin^3{\theta}$ y $\cos{\theta} = 4 \cos^3{\theta} - 3 \cos{\theta}$ obtenemos $$|\psi_3\rangle = \sin{3 \theta}|A\rangle + \cos{3 \theta}|B\rangle$$

## Clase 6:

$\textbf{9.- La implementación de QFT se probó preparando el estado de Fourier $|\widetilde{5}\rangle$ para el cual $\text{QFT}^{\dagger}|\widetilde{5}\rangle= |101\rangle$. Intenta encontrar el estado $|a\rangle$ tal que $\text{QFT}^{\dagger}|a\rangle = |100\rangle$.}$

Queremos encontrar el estado $|\tilde a\rangle$ tal que $QFT^\dagger|\tilde a\rangle = |a\rangle = |100\rangle$, i.e. $|\tilde a\rangle = QFT|a\rangle = QFT|100\rangle$.

Pero $|100\rangle$ en binario es $4$, por lo que el estado que buscamos es $|\tilde 4\rangle$.

Implementamos el circuito para aplicar QFT a tres qubits en el estado $|100\rangle$ y veremos que el estado resultante es $|\tilde 4\rangle$, verificaremos esto preparando el estado $|\tilde 4\rangle$ y aplicando el proceso inverso y al medir obtener $|100\rangle$

```python
from qiskit import QuantumCircuit, transpile, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import numpy as np
```

```python

# usaremos 3 qubits
qc = QuantumCircuit(3)

# inicializamos el estado 4: 100
qc.x(2)

# aplicamos H al qubit mas significativo
qc.h(2)
qc.draw('mpl')
```

```python
#si el qubit q_1 esta en 1 aplicamos ROT al qubit q_2
qc.cp(np.pi/2,1,2)

# si el qubit q_0 esta en 1 aplocamos ROT a q_2
qc.cp(np.pi/4,0,2)

qc.draw('mpl')
```

```python
# repetimos el proceso para los qubits q_1 y q_0
qc.h(1)
qc.cp(np.pi/2,0,1)

qc.h(0)

qc.draw('mpl')
```

```python
# we swap qubits q_0 and q_2
qc.swap(0,2)

qc.draw('mpl')
```

```python
backend = AerSimulator()
qc_init = qc.copy()
qc_init.save_statevector()
statevector = backend.run(qc_init).result().get_statevector()
plot_bloch_multivector(statevector)
#statevector.draw('latex')
```

Este es el estado $|\tilde 4 \rangle$. Para verificarlo, inicializamos dicho estaado y aplicamos $QFT^\dagger$ para obtener $|4\rangle = |100\rangle$:

```python
# ahora preparamos el estado |\tilde 4\rangle:
nqubits = 3
num = 4
qc1 = QuantumCircuit(nqubits)
for qubit in range(nqubits):
    qc1.h(qubit)
qc1.p(num*np.pi/4,0)
qc1.p(num*np.pi/2,1)
qc1.p(num*np.pi,2)

qc1.draw('mpl')
```

```python
# creamos el proceso inverso del circuito
inv_qc = qc.inverse()
inv_qc.x(2)
inv_qc.draw('mpl')
```

```python
q_c = qc1.compose(inv_qc)
q_c.measure_all()
q_c.draw('mpl')
```

```python
backend = AerSimulator()
#qc_init = qc.copy()
#qc_init.save_statevector()
fc=transpile(q_c,backend)
counts = backend.run(fc,shots=1000000).result().get_counts()
plot_histogram(counts)
#statevector1 = backend.run(qc_init).result().get_statevector()
#plot_bloch_multivector(statevector1)
#statevector.draw('latex')
```

Lo cual verifica que el estado $|\tilde a\rangle$ que buscamos es $|\tilde 4\rangle$.
